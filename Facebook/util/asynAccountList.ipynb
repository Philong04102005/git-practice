```python
import pandas as pd
```

```python
df_1 = pd.read_csv(r'C:\PhiLong\Revoland\Revoland_AI-Automation\Seeding\Facebook\data\account\newAccount.csv',)
```

```python
df_2 = pd.read_csv(r'C:\PhiLong\Revoland\Revoland_AI-Automation\Seeding\Facebook\data\account\account.csv')
```

```python
df_pre_final = pd.merge(df_1, df_2, on=['username', 'password'], how='left')
df_pre_final.drop("name_y", axis=1, inplace=True)
df_pre_final.rename(columns={"name_x": "name"}, inplace=True)
df_pre_final
```

```python
df_old = pd.read_csv(r"C:\PhiLong\Revoland\Revoland_AI-Automation\Seeding\Facebook\data\account\account_checked.csv")
df_old
```

```python
df_final = pd.merge(df_pre_final, df_old, on=["acc_id", "username", "password"], how="left")
df_final.drop(columns=["name_y", "status_x"], inplace=True)
df_final.rename(columns={"status_y": "status"}, inplace=True)
df_final.rename(columns={"name_x": "name"}, inplace=True)
df_final
```

```python
df_final = df_final[["acc_id","username","password","name","status","TotalFriends"]]
df_final
```

```python
df_final.to_csv(r"C:\PhiLong\Revoland\Revoland_AI-Automation\Seeding\Facebook\data\account\account_checked.csv", index=False)
```
