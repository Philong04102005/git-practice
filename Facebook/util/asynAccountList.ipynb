```python
import pandas as pd
```

```python
account_checked = pd.read_csv(f"../data/account/account_checked.csv")
```

```python
account_new = pd.read_csv(f"../data/account/account_new.csv")
```

```python
account_new["TotalFriend_2025-10-31"].fillna(0, inplace=True)
account_new.drop("TotalFriends_2025-10-31", axis=1, inplace= True)
account_new.rename(columns={"TotalFriend_2025-10-31": "TotalFriends_2025-10-29"}, inplace=True)
account_new
```

```python
account_checked.drop(columns=["TotalFriends", "FriendsGrow", "Message", "Post"], inplace=True)
account_checked
```

```python
df_pre_final = pd.concat([account_checked, account_new], ignore_index=True)
df_pre_final
```

```python
df_pre_final.drop_duplicates(subset=None, keep='first', inplace=False)
df_pre_final
```

```python
df_pre_final.rename(columns={"TotalFriends_2025-10-29": "TotalFriends"}, inplace=True)
df_pre_final
```

```python
list_val = [16, 405, 302, 97, 159, 87, 203, 375, 244, 319, 534]
list_index = [25, 27, 28, 31, 32, 33, 35, 36]
index = 0
for i in range(len(df_pre_final)):
    if int((df_pre_final.iloc[i, 0]).split("_")[1]) in list_index:
        df_pre_final.iloc[i, 5] = list_val[index]
        df_pre_final.iloc[i, 4] = True
        index += 1
df_pre_final
```

```python
df_good = df_pre_final[df_pre_final["TotalFriends"] > 0].reset_index(drop=True)
df_good
```
